```python
buffer = ""
cursor = {"row": 0, "column": 0}
text = []


def move_cursor_to_head():
    cursor["column"] = 0


def move_cursor_to_end():
    cursor["column"] = len(text[cursor["row"]])-1


def move_cursor_up_and_to_head():
    if is_there_a_char_to("above"):
        move_cursor_by_one_position_to("above")
    move_cursor_to_head()


def move_cursor_down_and_to_head():
    if is_there_a_char_to("below"):
        move_cursor_by_one_position_to("below")
    move_cursor_to_head()


def move_cursor_to_right():
    if is_there_a_char_to("right"):
        move_cursor_by_one_position_to("right")
    elif is_there_a_char_to("below"):
        move_cursor_by_one_position_to("below")
        move_cursor_to_head()
    else:
        pass


def move_cursor_to_left():
    if is_there_a_char_to("left"):
        move_cursor_by_one_position_to("left")
    elif is_there_a_char_to("above"):
        move_cursor_by_one_position_to("above")
        move_cursor_to_end()
    else:
        pass


def delete_char_on_cursor():
    buf = ""
    if is_there_a_char_on_cursor():
        text[cursor["row"]].pop(cursor["column"])
    elif is_there_a_char_to("below"):
        buf = text.pop(cursor["row"]+1)
        text[cursor["row"]].pop(-1)
        text[cursor["row"]].extend(buf)
    else:
        pass


def delete_with_yanking():
    global buffer
    if is_there_a_char_on_cursor():
        buffer = text[cursor["row"]][cursor["column"]:-1]
        del text[cursor["row"]][cursor["column"]:-1]
        move_cursor_to_end()
    else:
        delete_char_on_cursor()
        buffer = "\n"
    assert buffer != "", "The buffer is empty!"


def paste_buffer():
    buf = []
    if buffer == "":
        pass
    elif buffer == "\n":
        buf = text[cursor["row"]][cursor["column"]:]
        del text[cursor["row"]][cursor["column"]:-1]
        text.insert(cursor["row"]+1, buf)
        move_cursor_by_one_position_to("below")
        move_cursor_to_head()
    else:
        text[cursor["row"]][cursor["column"]:cursor["column"]] = buffer
        cursor["column"] += len(buffer)


def is_there_a_char_to(direction):
    if direction == "right":
        if cursor["column"] == len(text[cursor["row"]])-1:
            return False
        else:
            return True
    elif direction == "left":
        if cursor["column"] == 0:
            return False
        else:
            return True
    elif direction == "above":
        if cursor["row"] == 0:
            return False
        else:
            return True
    elif direction == "below":
        if len(text)-2 == cursor["row"]:
            return False
        else:
            return True


def move_cursor_by_one_position_to(direction):
    if direction == "above":
        cursor["row"] -= 1
    elif direction == "below":
        cursor["row"] += 1
    elif direction == "left":
        cursor["column"] -= 1
    elif direction == "right":
        cursor["column"] += 1
    else:
        pass


def is_there_a_char_on_cursor():
    cursor_char = text[cursor["row"]][cursor["column"]]
    if cursor_char != "\n" and cursor_char != "EOF":
        return True
    else:
        return False


def read_until_eof():
    line = ""
    while True:
        line = input()
        if line == "END_OF_TEXT":
            text.append(["EOF"])
            break
        line = list(line)
        line.append("\n")
        text.append(line)


def execute_command(command):
    if command == "a":
        move_cursor_to_head()
    elif command == "e":
        move_cursor_to_end()
    elif command == "p":
        move_cursor_up_and_to_head()
    elif command == "n":
        move_cursor_down_and_to_head()
    elif command == "f":
        move_cursor_to_right()
    elif command == "b":
        move_cursor_to_left()
    elif command == "d":
        delete_char_on_cursor()
    elif command == "k":
        delete_with_yanking()
    elif command == "y":
        paste_buffer()
    else:
        pass

def main():
    read_until_eof()

    while True:
        command = input()
        if command == "-":
            break
        execute_command(command)

    while True:
        line = text.pop(0)
        if not line or line[0] == "EOF":
          break
        line.pop(-1)
        line = "".join(line)
        print(line)

if __name__ == "__main__":
    main()
```
